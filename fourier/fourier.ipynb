```python
import numpy as np
import matplotlib.pyplot as plt
import math

def a_n(n):
    if n % 2 == 0:
        return 0
    return -4/((n**2)*np.pi**2)

def b_n(n):
    if n % 2 == 0:
        return -2/(n*np.pi)
    return 2/(n*np.pi)  

f = 236*10 # Hz
p = 1/f
a_0 = 0.2#236/100

t = np.arange(0, 3*p, 0.0000001)

omega = 2*np.pi*t*f

z = np.cos(omega)

y = a_0 

for i in range(1, 1000):
    y += a_n(i)*np.cos(i*omega) + b_n(i)*np.sin(i*omega)

Fs = 5 # Sample rate
n = len(y) # length of the signal
k = np.arange(n)[:30]
T = n/Fs
frq = k/T # two sides frequency range

y_ft = (np.fft.fft(y)/len(y))[:30]



fig, ax = plt.subplots(2,1)
ax[0].plot(t, y)
ax[0].plot(t, z)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(y_ft), 'r')
ax[1].set_xlabel('Freq')
ax[1].set_ylabel('y freq')

```
